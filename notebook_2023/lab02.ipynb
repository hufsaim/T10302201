- 일반적인 complex exponential signal을 정의하여, 다양한 C와 alpha에 대한 signal을 시각화하여 봅니다.


```python
def xn(n,C,alpha):
    xn = np.zeros(n.shape)
    xn = C * np.power(alpha,n)
    return xn
```

```python
N=20
n = np.arange(-N,N+1)

absC = -0.5
absA = 0.95
phaseC = (np.pi)*(1.2)
phaseA = (np.pi)*(2/12)

C = absC * np.exp(phaseC*1j)
A = absA * np.exp(phaseA*1j)
```

```python
k = xn(n, C, A)

plt.stem(n,np.real(k))
plt.xlabel('n')
plt.title('real part')
plt.grid('on')
plt.show()

plt.stem(n,np.imag(k))
plt.xlabel('n')
plt.title('imaginary part')
plt.grid('on')
plt.show()

plt.stem(n,np.abs(k))
plt.xlabel('n')
plt.title('magnitude')
plt.grid('on')
plt.show()

plt.stem(n,np.angle(k))
plt.xlabel('n')
plt.title('phase')
plt.grid('on')
plt.show()
```

- exp(jwn)에 대해 다양한 w에 대해 fundamental period가 어떻게 변하는지 살펴봅시다.


```python
def xn(n,w0):
    xn = np.zeros(n.shape)
    xn = np.exp(1j*w0*n)
    return xn
```

```python
N = 16
n = np.arange(-N,N)
```

```python
w0 = (np.pi)*(2/16)
k = xn(n,w0)
plt.stem(n,np.real(k), markerfmt='C0.')
plt.xlabel('n')
plt.title('real part')
plt.grid('on')
plt.show()

plt.stem(n,np.imag(k), markerfmt='C0.')
plt.xlabel('n')
plt.title('imaginary part')
plt.grid('on')
plt.show()

plt.stem(n,np.abs(k), markerfmt='C0.')
plt.xlabel('n')
plt.title('magnitude')
plt.grid('on')
plt.show()

plt.stem(n,np.angle(k), markerfmt='C0.')
plt.xlabel('n')
plt.title('phase')
plt.grid('on')
plt.show()
```

```python
w0 = (np.pi)*(2/16)
k = xn(n,w0+1*(np.pi))
plt.stem(n,np.real(k), markerfmt='C0.')
plt.xlabel('n')
plt.title('real part')
plt.grid('on')
plt.show()

plt.stem(n,np.imag(k), markerfmt='C0.')
plt.xlabel('n')
plt.title('imaginary part')
plt.grid('on')
plt.show()

plt.stem(n,np.abs(k), markerfmt='C0.')
plt.xlabel('n')
plt.title('magnitude')
plt.grid('on')
plt.show()

plt.stem(n,np.angle(k), markerfmt='C0.')
plt.xlabel('n')
plt.title('phase')
plt.grid('on')
plt.show()
```
