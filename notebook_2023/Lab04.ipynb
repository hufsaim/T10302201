# Fourier series representation and LTI system
- Fourier series로 표현하는 input signal에 대한 LTI system의 frequency response에서 phase가 가지는 의미를 이해하자.

```python
import numpy as np
import matplotlib.pyplot as plt
```

# 입력신호 정의하기
- N=200을 fundamental period로 하는 periodic signal
- w0 = 2pi/N
- w1, w2, w4에 대해 Fourier series coefficient를 정의하여 x[n]을 생성
- x[n]의 real part를 plot

```python
n = np.arange(0,200,1)
w0 = 2*np.pi/200

a1 = 0.75
x1 = a1 * np.exp(1j*1*w0*n)

a2 = 0.5
x2 = a2 * np.exp(1j*2*w0*n)

a4 = 0.25
x4 = a4 * np.exp(1j*4*w0*n)

x = x1 + x2 + x4


plt.subplot(411)
plt.plot(np.concatenate((np.real(x1),np.real(x1),np.real(x1)),axis=0))
plt.title('1st harmonic component')
plt.grid('on')

plt.subplot(412)
plt.plot(np.concatenate((np.real(x2),np.real(x2),np.real(x2)),axis=0))
plt.title('2nd harmonic component')
plt.grid('on')

plt.subplot(413)
plt.plot(np.concatenate((np.real(x4),np.real(x4),np.real(x4)),axis=0))
plt.title('4th harmonic component')
plt.grid('on')

plt.subplot(414)
plt.plot(np.concatenate((np.real(x),np.real(x),np.real(x)),axis=0),'r')
plt.title('real part of x[n]')
plt.grid('on')

plt.tight_layout()
plt.show()
```

# Constant phase response
- System의 frequency response가 모든 frequency에 대해 constant인 경우를 가정해보자.
- 적절한 constant C를 정해, 이로 인한 출력의 real part를 plot해보자.
- Magntidue는 변하지 않는 시스템. 즉, All pass filter라고 하자.

```python
C = 1.4

H1 = 1*np.exp(1j*C)
H2 = 1*np.exp(1j*C)
H4 = 1*np.exp(1j*C)

y1 = H1*x1
y2 = H2*x2
y4 = H4*x4

y = y1 + y2 + y4
y_constantphase = y.copy()

plt.subplot(411)
plt.plot(np.concatenate((np.real(y1),np.real(y1),np.real(y1)),axis=0))
plt.title('1st harmonic component')
plt.grid('on')

plt.subplot(412)
plt.plot(np.concatenate((np.real(y2),np.real(y2),np.real(y2)),axis=0))
plt.title('2nd harmonic component')
plt.grid('on')

plt.subplot(413)
plt.plot(np.concatenate((np.real(y4),np.real(y4),np.real(y4)),axis=0))
plt.title('3rd harmonic component')
plt.grid('on')

plt.subplot(414)
plt.plot(np.concatenate((np.real(y),np.real(y),np.real(y)),axis=0),'r')
plt.title('real part of y[n]')
plt.grid('on')

plt.tight_layout()
plt.show()
```

## Linear phase response
- frequency에 비례하여 phase가 변하는 경우를 가정해보자.
- 적절한 상수 A를 정하고, harmonic frequency에 비례하여 phase가 더해지도록 하자.
- 출력 y[n]의 real part를 관찰해보자.

```python
A = 1.4

H1 = 1*np.exp(1j*A*1)
H2 = 1*np.exp(1j*A*2)
H4 = 1*np.exp(1j*A*4)

y1 = H1*x1
y2 = H2*x2
y4 = H4*x4

y = y1 + y2 + y4
y_linearphase = y.copy()

plt.subplot(411)
plt.plot(np.concatenate((np.real(y1),np.real(y1),np.real(y1)),axis=0))
plt.title('1st harmonic component')
plt.grid('on')

plt.subplot(412)
plt.plot(np.concatenate((np.real(y2),np.real(y2),np.real(y2)),axis=0))
plt.title('2nd harmonic component')
plt.grid('on')

plt.subplot(413)
plt.plot(np.concatenate((np.real(y4),np.real(y4),np.real(y4)),axis=0))
plt.title('3rd harmonic component')
plt.grid('on')

plt.subplot(414)
plt.plot(np.concatenate((np.real(y),np.real(y),np.real(y)),axis=0),'r')
plt.title('real part of y[n]')
plt.grid('on')

plt.tight_layout()
plt.show()
```

## Linear phase vs. Constant phase
- 두 phase response의 출력을 비교하여 linear phase response의 필요성을 확인.

```python
plt.subplot(311)
plt.plot(np.concatenate((np.real(x),np.real(x),np.real(x)),axis=0),'r')
plt.title('real part of input x[n]')
plt.grid('on')

plt.subplot(312)
plt.plot(np.concatenate((np.real(y_constantphase),np.real(y_constantphase),np.real(y_constantphase)),axis=0),'r')
plt.title('real part of output of y[n] with constant phase system')
plt.grid('on')

plt.subplot(313)
plt.plot(np.concatenate((np.real(y_linearphase),np.real(y_linearphase),np.real(y_linearphase)),axis=0),'r')
plt.title('real part of output of y[n] with linear phase system')
plt.grid('on')


plt.tight_layout()
plt.show()
```
