```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.io import wavfile
from scipy import signal
from IPython.display import Audio
```

```python
fs,d = wavfile.read('hufswav999.wav')
c = d[:,0] + 1j*d[:,1]
t = np.arange(0,len(c))/fs
freq = np.fft.fftfreq(len(t),1/fs)
```

```python
plt.figure(figsize=(16,9))
plt.subplot(211)
plt.plot(t,np.real(c))
plt.title('real part of x[n]')
plt.grid('on')
plt.subplot(212)
plt.plot(t,np.imag(c))
plt.title('imaginary part of x[n]')
plt.grid('on')
plt.show()

Audio(np.real(c),rate=fs)
```

```python
plt.figure(figsize=(16,9))
plt.plot(freq,np.abs(np.fft.fft(c)))
plt.xlabel('Hz')
plt.title('magnitude of X[w]')
plt.grid('on')
plt.show()
```

# 학생1의 음성 복원하기
- frequency spectrum을 관찰하여 첫 번째 학생의 음성을 복원하기
- fr, wc에 적절한 값을 유추하여 신호를 복원하고, 음성을 확인하기
- 복원된 신호는 경우에 따라 추가적은 처리(post-processing)이 필요할 수도 있으니 음성을 확인하고 필요시 적절한 프로세싱을 추가해보기

```python
fr = -7200
wc = 3000

wr = wc # frequency for lowpass filtering
hr = np.exp(1j*2*np.pi*fr*t)
HR = np.ones_like(c)
HR[freq<-np.abs(wr)]=0
HR[freq>np.abs(wr)]=0
rd = c*hr
RD = np.fft.fft(rd)
plt.plot(freq,np.abs(RD))
RDL = RD*HR
plt.plot(freq,np.abs(RDL))
plt.show()

s1 = np.fft.ifft(RDL)
s1 = np.real(s1)

## post processing
# s1 = ...
```

- 학생1의 음성이 제대로 복원되는지 확인

```python
Audio(s1,rate=fs)
```
