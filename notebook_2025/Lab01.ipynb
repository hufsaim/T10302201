```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.io import wavfile
from scipy import signal
from IPython.display import Audio
```

```python
f1 = '21.wav'
f2 = '22.wav'
f3 = '23.wav'
```

```python
fs, d = wavfile.read(f1)
print(f"sampling rate: {fs} Hz")
print(f"shape of data: {d.shape}")
if len(d.shape)==2:
    wav1 = np.array(d[:,0])
elif len(d.shape)==1:
    wav1 = np.array(d)
L = len(d)
print(f"duration of data: {L/fs} s")
print(f"samples of data: {L} samples")
plt.plot(wav1)
plt.xlabel('n')
plt.grid('on')
plt.show()
#Audio(wav1,rate=fs)
```

```python
fs, d = wavfile.read(f2)
print(f"sampling rate: {fs} Hz")
print(f"shape of data: {d.shape}")
if len(d.shape)==2:
    wav2 = np.array(d[:,0])
elif len(d.shape)==1:
    wav2 = np.array(d)
#Audio(wav2,rate=fs)
```

```python
fs, d = wavfile.read(f3)
print(f"sampling rate: {fs} Hz")
print(f"shape of data: {d.shape}")
if len(d.shape)==2:
    wav3 = np.array(d[:,0])
elif len(d.shape)==1:
    wav3 = np.array(d)
#Audio(wav3,rate=fs)
```

```python
plt.subplot(311)
plt.plot(wav1)
plt.xlabel('n')
plt.grid('on')
plt.subplot(312)
plt.plot(wav2)
plt.xlabel('n')
plt.grid('on')
plt.subplot(313)
plt.plot(wav3)
plt.xlabel('n')
plt.grid('on')
plt.tight_layout()
plt.show()
```

```python
# total energy

E1 = np.sum(wav1**2)
E2 = np.sum(wav2**2)
E3 = np.sum(wav3**2)

print("total energy:")
print(E1, E2, E3)
```

```python
# averaged power

P1 = E1/len(wav1)
P2 = E2/len(wav2)
P3 = E3/len(wav3)

print("length of signals(sequences):")
print(len(wav1),len(wav2),len(wav3))
print("time-averaged power:")
print(P1, P2, P3)
```

```python
# sum of signals
L_min = np.min(np.array([len(wav1),len(wav2),len(wav3)]))
wav1c = wav1[:L_min]
wav2c = wav2[:L_min]
wav3c = wav3[:L_min]

mysig = (wav1c) + (wav2c) + (wav3c)
plt.plot(mysig)
plt.xlabel('n')
plt.grid('on')
plt.show()
#Audio(mysig,rate=fs)
```

```python
# time scaling
mysig = wav2[::3]
plt.plot(mysig)
plt.xlabel('n')
plt.grid('on')
plt.show()
#Audio(mysig,rate=fs)
```

```python
# time reversal
mysig = wav2[::-1]
plt.plot(mysig)
plt.xlabel('n')
plt.grid('on')
plt.show()
#Audio(mysig,rate=fs)
```

```python
# even part of signal
mysig_e = 0.5*(wav2+wav2[::-1])
plt.plot(mysig_e)
plt.xlabel('n')
plt.grid('on')
plt.show()
#Audio(mysig_e,rate=fs)
```

```python
# odd part of signal
mysig_o = 0.5*(wav2-wav2[::-1])
plt.plot(mysig_o)
plt.xlabel('n')
plt.grid('on')
plt.show()
#Audio(mysig_o,rate=fs)
```

```python
# sum of even and odd parts
mysig = mysig_e + mysig_o
plt.plot(mysig)
plt.xlabel('n')
plt.grid('on')
plt.show()
#Audio(mysig,rate=fs)
```
