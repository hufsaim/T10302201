### Wav파일 불러오기, sampling rate확인, 파형의 시각화

```python
from scipy.io import wavfile
from matplotlib import pyplot as plt
import numpy as np
from IPython.display import Audio
```

```python
d = wavfile.read('mywav.wav') # 각자의 wav파일을 read할 수 있도록 합니다.
```

```python
print(f"sampling rate: {d[0]} Hz")
```

```python
mywav = d[1]
print(mywav.shape)
```

```python
t = np.arange(0,len(mywav))/d[0]
plt.plot(t,mywav)
plt.xlabel('time (sec)')
plt.ylabel('signal intensity')
plt.grid('on')
plt.show()
```

```python
Audio(mywav,rate=d[0])
```
