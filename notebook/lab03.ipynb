```python
import numpy as np
import matplotlib.pyplot as plt
```

![square wave and its Fourier coefficients]('./squarewave.png')

## Square waves for different T1 and T

```python
T1 = 1.
T = 8. # should be larger than 2*T1

t = np.arange(-128,128,.01)
x = np.zeros_like(t).astype(np.float64)
r = x.copy()
r[(t%T)<0.0025] = 1
x[np.abs(t)<T1] = 1
x = np.convolve(r,x,'same')

plt.figure(figsize=(16,9))
plt.plot(t,x)
plt.axis([-12,12,-.2,1.2])
plt.grid('on')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title('Square Wave')
plt.show()
```

## Fourier Series Coefficients of Square Wave

```python
N = 64
k = np.arange(-N,N+1)

ak = np.zeros_like(k).astype(np.float64)
for n in range(len(k)):
    if k[n] == 0:
        ak[n] = 2*T1/T
    else:
        ak[n] =  np.sin(k[n]*(2*np.pi/T)*T1)/(k[n]*np.pi)

plt.figure(figsize=(16,9))
plt.stem(k,ak)
plt.xlabel('k')
plt.ylabel('ak')
plt.grid('on')
plt.title('Fourier Coefficients of Square Wave')
plt.show()
```

## Fourier Series Approximation of Square Wave for different N

```python
N = 2
t = np.arange(-128,128,.01)
x_N = np.zeros_like(t).astype(np.complex64)
for n in np.arange(-N,N+1):
    if n == 0:
        x_N += 2*T1/T
    else:
        x_N += 1*(np.sin(n*(2*np.pi/T)*T1)/(n*np.pi)) * np.exp(1j * n * ((2*np.pi)/T) * t) #  += ( ak ) * exp(j k w0 t)

E_N = np.sum( (np.abs(x - x_N)**2)[np.abs(t)<(T/2)] )

plt.figure(figsize=(16,9))
plt.plot(t,np.real(x_N))
plt.plot(t,np.real(x),'r:')
plt.plot(t,np.zeros_like(t),':')
plt.plot(t,1.1*np.ones_like(t),':')
plt.axis([-12, 12, -1.5, 1.5])
plt.grid('on')
plt.xlabel('t')
plt.ylabel('x_N(t)')
plt.title(f'Fourier Series Approximation (N ={N}), EN={E_N}')
plt.show()
```
