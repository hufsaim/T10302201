```python
import numpy as np
import matplotlib.pyplot as plt
```

- Impulse signal의 Fourier Transform은 모든 주파수성분이 같은 값을 가진다. 
    - F.T. of impulse train = 1
- 시뮬레이션을 통해 이에 대한 직관적인 이해를 가져보자.
- 충분히 작은 w0를 정의하자. (fundamental period T가 아주 긴 경우)

```python
t = np.arange(-100,100,.001)
w0 = 0.005
```

- w0의 +-k배에 해당하는 exp(ikw0t)를 계속 더해 나가면서 신호의 모양이 어떻게 바뀌는 지 살펴보자.

```python
N = 0
x_ = w0*np.ones_like(t).astype(np.complex64)
for n in range(N):
    w = (n+1)*w0
    x_ += w0*(np.exp((1j)*2*np.pi*w*t) + np.exp((1j)*2*np.pi*(-w)*t))
plt.plot(t,np.abs(x_))
plt.xlabel('time')
plt.axis([-100,100,-.1,1.5])
plt.grid('on')
plt.show()
```
